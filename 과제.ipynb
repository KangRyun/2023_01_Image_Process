```python
### Packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
### 출력 영상 크기
plt.rcParams["figure.figsize"] = (16,9)
### 한글 표시
plt.rcParams['font.family'] = "Gulim" # 'AppleGothic' in mac
```

```python
### 영상 읽기
img_raw = cv2.imread(r"C:\Users\kr937\Desktop\chicken.jpg")
```

```python
img_raw = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
```

```python
### 영상 출력 - grayscale
plt.imshow(img_raw)
plt.show()
```

```python
### 영상에 가우시안 잡음(noise) 추가
noise = np.random.normal(0,30, img_raw.shape)
img_noise_Gauss = np.clip(img_raw+noise, 0, 255).astype('uint8')

### 가우시안 필터
img_Gauss = cv2.GaussianBlur(img_noise_Gauss, (7, 7), 0)

### 양방향 필터
img_bF = cv2.bilateralFilter(img_noise_Gauss, -1, 50, 10)
```

```python
### 밝게 - alpha = 1, beta = 32
img_1 = cv2.convertScaleAbs(img_bF, alpha = 1, beta = 50)
```

```python
### 영상 출력
img_rgb = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
