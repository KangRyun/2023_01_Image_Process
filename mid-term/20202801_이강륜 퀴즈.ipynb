```python
kernel = np.ones((11,11), np.uint8)

### 팽창 연산
img_binary_inv_d = cv2.dilate(img_binary_inv, kernel)
```

```python
### 구조요소(커널)
kernel = np.ones((3,3), np.uint8)

### 침식 연산
img_binary_inv_e = cv2.erode(img_binary_inv, kernel)

### 열림 연산
img_binary_inv_open = cv2.morphologyEx(img_binary_inv, cv2.MORPH_OPEN, kernel)

### 닫힘 연산
img_binary_inv_c = cv2.morphologyEx(img_binary_inv, cv2.MORPH_CLOSE, kernel)


### 영상 출력
titles = ['원래영상', '침식연산', '팽창연산', '열림연산', '닫힘연산']
images = [img_binary_inv, img_binary_inv_e, img_binary_inv_d, img_binary_inv_open, img_binary_inv_c]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(2,3,i+1),plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
# 침식
img_e = cv2.bitwise_and(img_out, cv2.cvtColor(img_binary_inv_e, cv2.COLOR_GRAY2BGR))

# 팽창
img_d = cv2.bitwise_and(img_out, cv2.cvtColor(img_binary_inv_d, cv2.COLOR_GRAY2BGR))

# 열림
img_o = cv2.bitwise_and(img_out, cv2.cvtColor(img_binary_inv_open, cv2.COLOR_GRAY2BGR))

# 닫힘
img_c = cv2.bitwise_and(img_out, cv2.cvtColor(img_binary_inv_c, cv2.COLOR_GRAY2BGR))

### 영상 출력
titles = ['원래영상', '침식연산', '팽창연산', '열림연산', '닫힘연산']
images = [img_bp, img_e, img_d, img_o, img_c]
for i in range(len(images)):
    img_rgb = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
    plt.subplot(2,3,i+1),plt.imshow(img_rgb)
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
