## 11주차 - 형태학 처리


### ▶ 이진 영상의 형태학 기반 연산 - 개념
#### ■ 형태학(Morphology, 모폴로지)
영상의 기하학적 구조를 분석하거나 처리하는 기법   
객체의 모양(형태)을 다루는 것   
경계, 골격구조, convex-hull 등의 모양 관련 특징을 추출하기 위하여 형태를 표현 또는 기술   
(description)하는 유용한 도구   
응용 분야   
형태학적 필터링, 세선화(thinning), 가지치기(pruning), 전/후처리 등   
이진 영상(binary image)과 그레이스케일 영상(grey-scale image)에 대한 형태학적 처리가 가능   

```python
### Packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
### 출력 영상 크기
plt.rcParams["figure.figsize"] = (16,9)
### 한글 표시
plt.rcParams['font.family'] = "Gulim" 
```

```python
### 영상 읽기 - Grayscale
img_lena_gray= cv2.imread(r'C:\Users\kr937\Desktop\drive\2023\Image Process\lena.png', 0)
### 구조요소(커널)
kernel = np.ones((3, 3), np.uint8)
### 침식 연산
img_erode = cv2.erode(img_lena_gray, kernel)
### 팽창 연산
img_dilate = cv2.dilate(img_lena_gray, kernel)
### 영상 출력
titles = ["입력 영상", "침식 영상", "팽창 영상"]
images = [img_lena_gray, img_erode, img_dilate]
for i in range(len(images)):
    plt.subplot(1, len(images), i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
