# ■ QR 코드
- QR 코드(영어: QR code, Quick Response code)는 컴퓨터가 만든 흑백 격자무늬 패턴 코드로, 정보를 나타내는 매트릭스 형식의 이차원 코드
- 한국, 일본, 중국, 영국, 미국 등에서 많이 사용
- 덴소 웨이브이 등록 상표 'Quick Response'에서 유래
- 숫자 외에 문자의 데이터를 저장
    - 숫자 최대 7,089자
    - 영문자와 한글은 최대 4,296자
    - 한자 1,817자
- https://www.qrcode.com/ko/
    - 버전은 1~40으로 구성
        - 버전1(21×21cell)로 시작하여 가로/세로 각각 4cell씩 늘어, 버전40(177×177cell)로 설정

```python
### 내용
msg = 'http://bigdata.hannam.ac.kr/'

### QR image 생성
img = qrcode.make(msg)

### png 파일로 저장
img.save("QR_BDA.png")
```

```python
### 내용
msg = '''박영호 통계학박사
조교수, 한남대학교 스마트융합대학 빅데이터응용학과
(34430) 대전광역시 대덕구 한남로 70 한남대학교 경상대학 4층 50417호
Tel : 042-629-7513
Fax : 0504-286-8205
Mobile : 010-9091-8205
E-mail : yhpark@hnu.kr'''
```

```python
### QR image 생성
img = qrcode.make(msg)
```

```python
### QR image 생성 - Advanced Usage
qr = qrcode.QRCode(
    version = 10, # size of the QR code: 1~40
    error_correction = qrcode.constants.ERROR_CORRECT_M, # About 15% or less errors can
    box_size = 10, # 박스(Box) 하나의 화소(pixel) 크기
    border = 4 # image border와 qrcode 사이 간격
)
qr.add_data(msg)
qr.make(fit = True) # QR 코드의 크기를 자동으로 결정
img = qr.make_image(fill_color = "black", black_color = "white")
```

```python
### png 파일로 저장
img.save("QR_PYH.png")
```

```python
### QR 출력
from matplotlib import pyplot as plt

plt.imshow(img, "gray")
plt.axis('off')
plt.show()
```
