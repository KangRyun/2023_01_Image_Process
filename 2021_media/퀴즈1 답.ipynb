3. hnu_logo.jpg 영상을 grayscale로 변환 후, 밝기가 200 이상인 화소를 0, 아니면 255로 이진화하시오.

```python
### 이진화 - grayscale
img_gray = cv2.cvtColor(img_logo, cv2.COLOR_BGR2GRAY)
ret, img_thresh = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY_INV)

### 영상 출력
plt.imshow(img_thresh, 'gray')
plt.axis('off')
plt.show()
```

4. hun_logo.jpg 영사에 가우시안 필터(3, 3)를 적용 후 3번의 처리를 수행하시오.

```python
### 가우시안 필터
img_GB = cv2.GaussianBlur(img_logo, (3, 3), 0)

### 이진화 - grayscale
img_gray = cv2.cvtColor(img_GB, cv2.COLOR_BGR2GRAY)
ret, img_thresh = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY_INV)

### 영상 출력
plt.imshow(img_thresh, 'gray')
plt.axis('off')
plt.show()
```
