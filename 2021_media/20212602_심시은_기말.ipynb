```python
### Packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
from easygui import fileopenbox
import os
```

```python
### 출력 영상 크기
plt.rcParams["figure.figsize"] = (16,9)
### 한글 표시
plt.rcParams['font.family'] = "Gulim"
```

```python
### 영상 읽기
img_noise = cv2.imread(r'C:\Users\sse88\Downloads\img\noise_lena.jpg')

### 잡음 제거
img_GB = cv2.GaussianBlur(img_noise, (3, 3), 0)

### 영상 출력
x1, x2, y1, y2 = [300, 331, 70, 101]
titles = ["원래 영상", "가우시안 필터"]
images = [img_noise, img_GB]
for i in range(len(images)):
    plt.subplot(1,2,i+1), plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
### PSNR - OpenCV
[round(cv2.PSNR(img_noise, images[i]), 2) for i in range(len(images))]
```

```python
# 저장할 파일명 및 확장자
out_file = "out_lena.jpg"
cv2.imwrite(out_file, img_GB)
```
