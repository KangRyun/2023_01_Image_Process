```python
### Packages
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
### 출력 영상 크기
plt.rcParams["figure.figsize"] = (16,9)
### 한글 표시
plt.rcParams['font.family'] = "Gulim" 
```

```python
### 영상 출력 함수
def fn_imshow(img, axis='off'):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    if axis=='off': plt.axis('off')
    plt.show()
```

```python
### 좌표 출력
def fn_mouse_callback(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        print("Double click, [ x , y ]:[", x ,",", y,"]")
```

```python
### 모자이크 처리 함수
def fn_mosaic (img, rect, size = 9) :
    x1, y1, x2, y2 = rect
    w = x2 - x1
    h = y2 - y1
    img_small = cv2.resize(img[y1:y2, x1:x2], (size, size))
    img[y1:y2, x1:x2] = cv2.resize(img_small, (w, h), interpolation = cv2.INTER_AREA)
    return img
```

```python
### 영상 읽기
img_raw = cv2.imread(r"C:\Users\kr937\Desktop\drive\2023\Image Process\KIH.jpg")
fn_imshow(img_raw)
```

```python
img_raw = cv2.resize(img_raw, dsize=(0,0), fx=0.5, fy=0.5,
                           interpolation=cv2.INTER_AREA)
```

```python
### 영상 출력 - Display an image in an OpenCV window
win_name = "Display window"
cv2.imshow(win_name, img_raw)
cv2.setMouseCallback(win_name, fn_mouse_callback)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
### 모자이크 처리
img_out = fn_mosaic(img_raw, [143 , 403, 444, 793])
fn_imshow(img_out)
```
